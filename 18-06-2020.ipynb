# Encode the categorical columns using different kinds of encoding techniques from pandas and scikit-learn

```python
merge_data=pd.get_dummies(df,columns=['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation','status'])
merge_data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # creating an object
merge_data=sc.fit_transform(merge_data)
merge_data
```

```python
merge_data=pd.DataFrame(merge_data)
merge_data
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

X=ColumnTransformer(transformers=[('Encoder',OneHotEncoder(),['status'])],remainder="passthrough")

df1=X.fit_transform(df)
df1=pd.DataFrame(df1)
df1=df1.iloc[:,0:2]
df1.columns=['status_placed','status_not placed']
df1=pd.concat([df,df1],axis=1)
df1=df1.drop('status',axis=1)
df1
```
